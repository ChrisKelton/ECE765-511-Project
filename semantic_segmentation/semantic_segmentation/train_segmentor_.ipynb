```python
from semantic_segmentation.train_segmentor import main as train_segmentor_main
from semantic_segmentation.train_backbone import main as train_backbone_main
from semantic_segmentation.multiple_trainings import main as train_multiple_main
from pathlib import Path
import torch
import os

data_path = Path("/content/drive/MyDrive/ECE765-PGM/Project/data/gtFine_trainvaltest")
n_workers = len(os.sched_getaffinity(0))
batch_size = 1
learning_rate = 1e-5
weight_decay = 3e-4
epochs = 50
epochs_to_test_val = 2
```

```python
train_fcn_backbone: bool = False

if train_fcn_backbone:
  fcn_output_path = Path("/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/FcnResNet")
  try:
    train_backbone_main(
        data_path=data_path,
        n_workers=n_workers,
        batch_size=batch_size,
        learning_rate=learning_rate,
        epochs=epochs,
        epochs_to_test_val=epochs_to_test_val,
        output_path=fcn_output_path,
        resize_size=None,
    )
  except Exception as e:
    print(f"\n\nError: {e}")
    torch.cuda.empty_cache()
```

```python
train_crf_rnn_fcn_backbone: bool = False

if train_crf_rnn_fcn_backbone:
  crf_fcn_output_path = Path("/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/CRF-RNN")
  try:
    train_segmentor_main(
        data_path=data_path,
        n_workers=n_workers,
        batch_size=batch_size,
        learning_rate=learning_rate,
        epochs=epochs,
        epochs_to_test_val=epochs_to_test_val,
        output_path=crf_fcn_output_path,
        resize_size=None,
    )
  except Exception as e:
    print(f"\n\nError: {e}")
    torch.cuda.empty_cache()
```

```python
train_crf_rnn_finetuned_backbone_frozen_layers: bool = True

if train_crf_rnn_finetuned_backbone_frozen_layers:
  model_name: str = "CRF-RNN--finetuned-backbone--frozen-layers"
  pretrained_backbone_path: Path = Path("/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/FcnResNet/FcnResNet/FcnResNet--47.pth")
  freeze_backbone_layers: bool = True
  crf_rnn_finetuned_backbone_frozen_layers_out_path = Path(f"/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/{model_name}")
  try:
    train_multiple_main(
        model_name=model_name,
        data_path=data_path,
        pretrained_backbone_path=pretrained_backbone_path,
        freeze_backbone_layers=freeze_backbone_layers,
        n_workers=n_workers,
        batch_size=batch_size,
        learning_rate=learning_rate,
        epochs=epochs,
        epochs_to_test_val=epochs_to_test_val,
        output_path=crf_rnn_finetuned_backbone_frozen_layers_out_path,
        center_crop_size=None,
        resize_size=None,
    )
  except Exception as e:
    print(f"\n\nError: {e}")
    torch.cuda.empty_cache()
```

```python
train_crf_rnn_finetuned_backbone_non_frozen_layers: bool = True

if train_crf_rnn_finetuned_backbone_non_frozen_layers:
  model_name: str = "CRF-RNN--finetuned-backbone--not-frozen-layers"
  pretrained_backbone_path: Path = Path("/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/FcnResNet/FcnResNet/FcnResNet--47.pth")
  freeze_backbone_layers: bool = True
  crf_rnn_finetuned_backbone_frozen_layers_out_path = Path(f"/content/drive/MyDrive/ECE765-PGM/Project/semantic_segmentation/semantic_segmentation/train/checkpoints/{model_name}")
  try:
    train_multiple_main(
        model_name=model_name,
        data_path=data_path,
        pretrained_backbone_path=pretrained_backbone_path,
        freeze_backbone_layers=freeze_backbone_layers,
        n_workers=n_workers,
        batch_size=batch_size,
        learning_rate=learning_rate,
        epochs=epochs,
        epochs_to_test_val=epochs_to_test_val,
        output_path=crf_rnn_finetuned_backbone_frozen_layers_out_path,
        center_crop_size=None,
        resize_size=None,
    )
  except Exception as e:
    print(f"\n\nError: {e}")
    torch.cuda.empty_cache()
```
